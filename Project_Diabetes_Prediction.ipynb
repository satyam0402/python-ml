STEP 1 : Importing dependecies and Dataset

```python
import pandas as pd
import numpy as np
# Standardisation of data
from sklearn.preprocessing import StandardScaler
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
# Model SVM
from sklearn.svm import SVC
# Model Evaluation
from sklearn.metrics import confusion_matrix, accuracy_score

```

Step 2 : Data Collection and Analysis

```python
# Loading dataset of the Diebaties csv file to pandas dataframe.
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
# if in some case we dont know what a specific function does then we can use the following to know more about that function :
#pd.read_csv? # Feature of google collab.
```

```python
# Printing first five rows of dataset using head().
diabetes_dataset.head()
# NOTE : Col : Outcome is label that tells if a pateint is Diabetic or not 1 or 0 respectively.
```

```python
# Number of rows and columns in dataset
diabetes_dataset.shape # (rows,columns)
```

```python
# Getting Statistical measure of the data
diabetes_dataset.describe()
```

```python
# Value count for Outcome
diabetes_dataset['Outcome'].value_counts() # 0 --> NON DIABETIC 1 --> Diabetic
```

```python
# Mean for Outcome
diabetes_dataset.groupby('Outcome').mean()
```

```python
# Seperating data and labels
X = diabetes_dataset.drop(columns = 'Outcome', axis = 1) # Here specify the axis = 1 when droping coloumn and if droping row then specify axis = 0
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

STEP 3 : Data Standardization

Why data standardization is required ?
We have different number of features like glucose, bloodpressure, bmi etc and there is difference in the values of all these features and it will be difficult for the ml model to predict so we have to standardise the data in a particular range so the ml model can predict well.

```python
# Creating variable sc to store StandardScaler function featires. Then we use fit() function of standard scaler to fit all X.
sc  = StandardScaler()
sc.fit(X)
```

```python
# Now transforming the data.
standardized_data = sc.transform(X) # Here we are using standard scaler function for fitting the inconsistent data and transforming it to a more specific range so that our ml model can predic well.
```

```python
print(standardized_data)
```

```python
X = standardized_data
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

STEP 4 : Split data into Training and Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X_train, X_test, Y_train, Y_test)
```

```python
print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of Y_train:", Y_train.shape)
print("Shape of Y_test:", Y_test.shape)
```

Step 6: Training the model

```python
classifier_model = SVC(kernel = 'linear')
```

```python
# Training the svm classifier.
classifier_model.fit(X_train, Y_train)
```

STEP 7 : Model Evaluation

```python
#Accuracy Score on the training data
X_train_prediction = classifier_model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
# Accuracy score on test data
X_test_prediction = classifier_model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data : ', test_data_accuracy)
```

Step 8 : Prediction System

```python
input_data = (1,126,60,0,0,30.1,0.349,47)


# Changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#Reshape the array as we predict for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# Standardize the input data
std_data = sc.transform(input_data_reshaped)
print(std_data)
```

```python
# Prediction
prediction = classifier_model.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
