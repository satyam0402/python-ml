Matplotlib :
  - Useful for making plots

```python
# Importing matlpotlib library
import matplotlib.pyplot as plt
```

```python
# importing numpy to get data for the plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
print(x)
```

```python
# sin
y = np.sin(x)
print(y)
```

```python
# cos
z = np.cos(x)
print(z)
```

```python
# sine wave
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin(x)')
plt.title('Sine Wave')
plt.show()
```

```python
# cose wave
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('cos(x)')
plt.title('Cosine Wave')
plt.show()
```

```python
# Parabol
x = np.linspace(-10,10,100)
y = x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parabol')
plt.show
```

```python
# sin and cos waves in one plot

x = np.linspace(-5,5,100)
plt.plot(x, np.sin(x),'r--')
plt.plot(x , np.cos(x), 'g')
plt.show
```

Bar Plot

```python
# fig = plt.figure()
# ax = plt.figure().add_axes([0,0,1,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
plt.figure().add_axes([0,0,1,1]).bar(langs,students)
plt.xlabel('Programming Languages')
plt.ylabel('Students good at that language')
plt.title('Bar Plot')
plt.show()
```

Pie chart

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
plt.figure().add_axes([0,0,1,1]).pie(students, labels = langs, autopct = '%1.1f%%')
plt.show()
```

3D Scatter Plot

```python

fig = plt.figure()
ax = plt.axes(projection='3d')
z = 10*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter3D(x, y, z, c=z, cmap='Greens')
plt.show()
```
