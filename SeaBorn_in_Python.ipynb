HOUSE_BOSTON DATASET from SKLEARN

```python
from sklearn.datasets import fetch_california_housing
```

```python
house_calafornia = fetch_california_housing()

house = pd.DataFrame(house_calafornia.data,columns=house_calafornia.feature_names)
```

```python
house.tail()
```

```python
house['Price'] =  house_calafornia.target
```

```python
house.head()
```

Distribution Plot : Tells what is the value/Magnitude of the price and what range has max value

```python
sns.displot(house['Price'])
```

```python
# distplaot gives probability
sns.distplot(house['Price'])
```

Correlation Plot
- Positive Correlation
- Negative Correlation

HEAT MAP

```python
correlation_heatmap = house.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation_heatmap,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
